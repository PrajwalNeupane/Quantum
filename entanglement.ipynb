```python
from qiskit import *
```

```python
circuit = QuantumCircuit(2,2)
```

```python
#quantum_register = QuantumRegister(2)
#classical_register = ClassicalRegister(2)
#circuit = QuantumCircuit(quantum_register, classical_register)
```

```python
circuit.draw()
```

```python
%matplotlib inline
```

```python
circuit.draw(output='mpl')
```

```python
circuit.h(0) # apply hadamard gate
circuit.draw(output='mpl')
```

```python
circuit.cx(0,1) # 0 => control bit, 1 => target bit
circuit.draw(output='mpl')
```

```python
circuit.measure([0,1],[0,1]) # apply a measurement on [0,1] qubits and store measurements in [0,1] cbits
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator') # get simulator
```

```python
result = execute(circuit, simulator).result() # run on simulator and get results
```

```python
from qiskit.visualization import plot_histogram
```

```python
plot_histogram(result.get_counts(circuit)) # show results in a histogram 
```
